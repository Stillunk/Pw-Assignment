```python
#Q2. What are the differences between multiprocessing and multithreading?
""" In multithreading(using threading module in python), multiple threads at the same time are generated by a single process. 
In multiprocessing(using multiprocessing module in python), multiple threads at the same time run across multiple cores.
Multiprocessing uses two or more CPUs to increase computing power,
whereas multithreading uses a single process with multiple code segments to increase computing power"""
```

```python
#Q3. Write a python code to create a process using the multiprocessing module.
import multiprocessing
def test():
    print("This is a process")
if __name__=="__main__" :
    m=multiprocessing.Process(target=test)
    print("This is my main")
    
    m.start()
    m.join()
    
```

```python
#Q4. What is a multiprocessing pool in python? Why is it used?
"""The Pool class in multiprocessing can handle an enormous number of processes. It allows you to run multiple jobs per process(due to its ability to queue the jobs).
The memory is allocated only to the executing processes, unlike the Process class, which allocates memory to all the processes."""
def sq(n):
    return n**2

if __name__=="__main__":
    with multiprocessing.Pool(processes=4) as pool:
        out=pool.map(sq,[1,2,3,4,5,6])
    print(out)
   
```

```python
#Q5. How can we create a pool of worker processes in python using the multiprocessing module?
"""You can create a pool of worker processes in Python using the multiprocessing module, specifically by utilizing the multiprocessing.Pool class."""
def sq(n):
    return n**2

if __name__=="__main__":
    with multiprocessing.Pool(processes=4) as pool:
        out=pool.map(sq,[1,2,3,4,5,6])
    print(out)
```

```python
#Q6. Write a python program to create 4 processes, each process should print a different number using the multiprocessing module in python.
import multiprocessing
def squ(index,value):
    print("value by process {} is {}".format(index,value[index]))
if __name__=="__main__":
    arr= [1,2,3,4]
    process= []
    for i in range(4):
        m=multiprocessing.Process(target=squ , args=(i,arr))
        process.append(m)
        m.start()
    for m in process:
        m.join()
```
