Q7. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5. Interpret the results.

Ans :





```python
import scipy.stats as stats

mean = 50
std_dev = 5

# Sample size (assuming it's large enough for the normal approximation)
sample_size = 100

# Z-score for 95% confidence level
Z = stats.norm.ppf(0.975)  # 0.975 because we want to consider both tails of the distribution

# Calculate the margin of error
margin_of_error = Z * (std_dev / (sample_size ** 0.5))

# Calculate the lower and upper bounds of the confidence interval
lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

print("95% Confidence Interval: ({:.2f}, {:.2f})".format(lower_bound, upper_bound))
```

Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is significantly effective at a 95% confidence level using a t-test.


```python
from scipy.stats import t

# Given values
sample_mean = 6
sample_std_dev = 2.5
sample_size = 50
population_mean = 0  # Null hypothesis: population mean is 0

# Calculate the t-statistic
t_statistic = (sample_mean - population_mean) / (sample_std_dev / (sample_size ** 0.5))

# Degrees of freedom
degrees_of_freedom = sample_size - 1

# Calculate the critical t-value for a two-tailed test at 95% confidence level
alpha = 0.05
t_critical = t.ppf(1 - alpha/2, degrees_of_freedom)

# Compare the t-statistic with the critical t-value
if abs(t_statistic) > t_critical:
    print("Reject the null hypothesis: The weight loss drug is significantly effective.")
else:
    print("Fail to reject the null hypothesis: There is not enough evidence to conclude that the weight loss drug is significantly effective.")

print("t-statistic:", t_statistic)
print("Critical t-value:", t_critical)
```

Q11. In a survey of 500 people, 65% reported being satisfied with their current job. Calculate the 95% confidence interval for the true proportion of people who are satisfied with their job.

```python
import scipy.stats as stats
import numpy as np

# Given values
sample_size = 500
sample_proportion = 0.65  # 65% reported being satisfied with their job

# Calculate the standard error
standard_error = np.sqrt((sample_proportion * (1 - sample_proportion)) / sample_size)

# Calculate the margin of error
margin_of_error = stats.norm.ppf(0.975) * standard_error  # 0.975 because it's a two-tailed test

# Calculate the confidence interval
lower_bound = sample_proportion - margin_of_error
upper_bound = sample_proportion + margin_of_error

print("95% Confidence Interval: ({:.4f}, {:.4f})".format(lower_bound, upper_bound))
```
