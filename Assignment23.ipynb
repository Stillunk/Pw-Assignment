```python
#Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
# plot a scatter plot.
"""Scatter plot also known as dot plot it represent data in the form of dot on graph."""
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
ply.scatter(x,y)
ply.xlabel("This is x data")
ply.ylabel("This is y data")
ply.title("Scatter plot")
ply.show()
```

```python
#Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
# Use the following data:
# import numpy as np
# For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
# For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
# For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
# For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
"""The subplot() function is used in data visualization libraries like Matplotlib, which is commonly used with Python, to create multiple plots within the same figure.
"""
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
ply.subplot(2,2,1)
ply.plot(x,y)
ply.title("Plot 1")

y1 = np.array([50, 20, 40, 20, 60, 70])
ply.subplot(2,2,2)
ply.plot(x,y1)
ply.title("Plot 2")

y2 = np.array([10, 20, 30, 40, 50, 60])
ply.subplot(2,2,3)
ply.plot(x,y2)
ply.title("Plot 3")

y3= np.array([200, 350, 250, 550, 450, 150])
ply.subplot(2,2,4)
ply.plot(x,y3)
ply.title("Plot 4")

ply.tight_layout()
ply.show()
```

```python
#Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
"""Bar plot is used to create bar graph nd represent data in bars"""
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
ply.bar(company,profit)
ply.xlabel("company")
ply.ylabel("profit")
ply.show()
ply.barh(company,profit)
ply.xlabel("company")
ply.ylabel("profit")
ply.show()
```

```python
#Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
ply.boxplot(box1,box2)
```
